```python
import sqlite3
import pandas as pd
import json
```

```python
#These Json Files contain the CNG Groupings 
dictnoun = json.load(open("noungroups.json"))
dictverb = json.load(open("verbgroups.json"))
```

```python
conn = sqlite3.connect("Extracted.db")
```

```python
df = pd.DataFrame(columns = ["CNG", "CNG_Group"])
```

```python
df
```

```python
z =0
for key in dictnoun.keys():
    for value in dictnoun[key]:
        df.loc[z] = [value, key]
        z += 1
        
```

```python
df.head()
```

```python
for key in dictverb.keys():
    for value in dictverb[key]:
        df.loc[z] = [(-1)*value, key]
        z += 1
        
```

```python
df
```

```python
df.to_sql("CNG&CNGG", conn)
```

```python
conn.close()
```
